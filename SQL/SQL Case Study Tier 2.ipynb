```python
import sqlite3
```

```python
query10 = """WITH s AS (SELECT f.name,
                               CASE WHEN m.memid = 0 THEN b.slots * f.guestcost
                               ELSE b.slots * f.membercost END AS cost
                        FROM Members AS m
                        LEFT JOIN Bookings AS b
                          ON m.memid = b.memid
                        LEFT JOIN Facilities AS f
                          ON b.facid = f.facid)
SELECT name, SUM(cost) AS revenue
FROM s
GROUP BY name
HAVING revenue > 1000;""" 


query11 = """SELECT m1.surname, 
                    m1.firstname, 
                    m2.firstname || ' ' || m2.surname AS recommendedby 
             FROM Members AS m1 
             INNER JOIN Members AS m2 
               ON m1.recommendedby = m2.memid 
             WHERE m1.recommendedby NOT LIKE "" 
             ORDER BY m1.surname, m1.firstname;"""

query12 = """SELECT DISTINCT f.name, 
                    m.firstname || ' ' || m.surname AS name, 
                    COUNT(b.bookid) OVER(PARTITION BY f.name, b.memid) AS member_usage
             FROM Bookings AS b
             LEFT JOIN Facilities AS f
               ON b.facid = f.facid
             LEFT JOIN Members AS m
               ON b.memid = m.memid
             WHERE b.memid <> 0;"""

query13 = """WITH s AS (SELECT bookid, 
                               STRFTIME('%m', starttime) AS month
                        FROM Bookings)
SELECT DISTINCT f.name, 
	   s.month,
       COUNT(b.bookid) OVER(PARTITION by f.name, s.month) AS monthly_usage
FROM Bookings AS b 
LEFT JOIN s
  ON b.bookid = s.bookid
LEFT JOIN Facilities AS f
  ON b.facid = f.facid
WHERE b.memid <> 0;
"""
```

```python
with sqlite3.connect('sqlite_db_pythonsqlite.db') as con:
    cur = con.cursor()
    for row in cur.execute(query10):
        print(row)
    
```

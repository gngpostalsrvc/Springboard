```python
#import the necessary packages
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
#load the data
df = pd.read_json('logins.json')

df.head()
```

The first order of business is to check whether there is any missing data and whether the datatypes are correct. 

```python
df.isna().sum()
```

```python
df.dtypes
```

With that out of the way, let's check whether the values are in chronological order. 

```python
df.sort_values(by='login_time')
```

They aren't, so we'll need to rearrange them before proceeding. 

```python
df = df.sort_values(by='login_time')

df.reset_index(drop=True, inplace=True)

df.head()
```

I defined a helper function to assign each login time to the appropriate 15 minute interval beginning with 8:15pm on January 1, 1970 and applied it the DataFrame.

```python
def rounder(date):
    mins = date.minute
    if 0 <= mins & mins <= 15:
         new_time = dt.datetime(date.year, date.month, date.day, date.hour, 15)
    elif 16 <= mins & mins <= 30:
        new_time = dt.datetime(date.year, date.month, date.day, date.hour, 30)
    elif 31 <= mins & mins <= 45:
        new_time = dt.datetime(date.year, date.month, date.day, date.hour, 45)
    else:
        new_time = dt.datetime(date.year, date.month, date.day, date.hour) + dt.timedelta(hours=1)
    return new_time
```

```python
df['login_by_interval'] = df['login_time'].apply(rounder)
```

I then pivoted the DataFrame in order to calculate the number of trips during each 15 minute interval. 

```python
df['number_of_trips'] = 1 

df_agg = df.pivot_table(values='number_of_trips', index='login_by_interval', aggfunc='sum').reset_index()
```

Finally, I plotted the data to get a better sense of the pattern of demand.  

```python
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(df_agg['login_by_interval'], df_agg['number_of_trips'], linewidth=.5, color='orange')
ax.set_title('Login Behavior January 1-April 13, 1970')
ax.set_ylabel('Number of Logins')
plt.show();
```

Although there is a regular pattern of spikes and dips, the number of logins appears relatively stationary over the three and a half month period covered by the data. Unfortunately, the current visualization is too crowded to get a sense of the daily changes in logins. For that information, we'll look at the data from a shorter period of time. 

```python
df_agg['day_of_week'] = df['login_by_interval'].dt.day_name()

df_agg.head()
```

```python
df_jan_week = df_agg[df_agg['login_by_interval'] < dt.datetime(1970, 1, 8)]
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
ax.plot(df_jan_week['login_by_interval'], df_jan_week['number_of_trips'], linewidth=.5, color='orange')
ax.set_xticklabels(['Friday', 'Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday'])
ax.set_title('Login Behavior January 1-8, 1970')
ax.set_ylabel('Number of Logins')
plt.show();
```

While logins spike at midday and in the evening every day of the week,  the relative size of these spikes varies depending on the day. On Sunday and Monday, the midday spike is larger than the evening one, but the opposite is true on Tuesday, Friday and Saturday. On Wednesday and Thursday, the spikes are roughly equal. Most likely, the size and distribution of these spikes correspond to the rhythm of life in the city: workers returning to the office from lunch, for example, may be responsible for the midday spikes, while people returning home from work or a night on the town could cause the evening spikes. Such a hypothesis would also help explain why the evening spikes are larger on Friday and Saturday since more people tend to go out in the evenings on the weekend.
